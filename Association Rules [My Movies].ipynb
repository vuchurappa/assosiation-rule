### Import all the required libraries 

```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori,association_rules

import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)
```

```python
df_movies = pd.read_csv("my_movies.csv")
df_movies.head()
```

```python
df_movies = df_movies[['SixthSense', 'Gladiator', 'LOTR1','HarryPotter1', 'Patriot', 'LOTR2', 'HarryPotter2', 'LOTR','Braveheart', 'GreenMile']]
```

```python
df_movies
```

### Apriori Algorithm 

```python
frequent_itemsets = apriori(df_movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
df_movies[df_movies["Gladiator"]==1].shape[0]/df_movies.shape[0]
```

```python
rules = association_rules(frequent_itemsets,metric = "lift", min_threshold=1.3)
rules
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold=1.1)
rules[rules["lift"]>1]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
